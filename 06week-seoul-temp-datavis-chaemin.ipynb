## 크리스마스의 기온 변화를 그래프로 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

result = []

for row in data:
    if row[4] != '':
        result.append(float(row[4]))

f.close()

plt.plot(result, 'r')
plt.xlabel('Day')
plt.ylabel('The highest temperature')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

mean = []
low = []
high = []

for row in data:
    if (row[2] != '') and (row[3] != '') and (row[4] != ''):
        if (row[0].split('-')[1] == '12') and (row[0].split('-')[2] == '25'):
            mean.append(float(row[2]))
            low.append(float(row[3]))
            high.append(float(row[4]))

f.close()

plt.figure(figsize=(10, 2))
plt.plot(mean, 'k')
plt.plot(low, 'b')
plt.plot(high, 'r')
plt.xlabel('Day')
plt.ylabel('The highest temperature')
plt.show()
```

# 기온 데이터를 히스토그램으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
jan = []
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[4] != '':
        if month == '01':
            jan.append(float(row[4]))
        elif month == '08':
            aug.append(float(row[4]))

plt.hist(jan, bins=100, color='b', label='Jan.')
plt.hist(aug, bins=100, color='r', label='Aug.')
plt.legend()
plt.show()
```

```python
# 기온 데이터를 상자 그림으로 표현하기 (5/6) , (6/6))
# 1월 일별 기온 데이터를 상자 그림으로 시각화하기
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

day = []

for i in range(31):
    day.append([])

for row in data:
    if row[4] != '':
        if row[0].split('-')[1] == '01':
            day[int(row[0].split('-')[2]) - 1].append(float(row[4]))
        
plt.style.use('ggplot')
plt.figure(figsize=(10, 5))
plt.boxplot(day, showfliers = False)
plt.show()
```

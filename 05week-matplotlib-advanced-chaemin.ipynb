```python
import matplotlib.pyplot as pit 
import numpy as np

x= np.linspace(-2 * np.pi, 2 *np.pi, 200) # 0에서 2*01까지 경계선을 포함해 199등분
pit. plot(x, np.sin(x) *10)
pit. plot (x, x**2)
pit. plot (x, 2 * x**2 - 3*x)
pit.grid(True)
pit.title('다양한 그래프')
pit. show()
```

# 난수와 돗수분포표

```python
import random
import matplotlib.pyplot as plt

dice = []

for i in range(5) :
    dice.append(random.randint(1,6))
    
plt.hist(dice,  bins=6)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ax1 = plt.subplot(2,1,1)
plt.plot(np.random.randint(1, 10, 10), 'yo-')
plt.title('a tale of 2 subplots')
plt.ylabel('damped osciliation')
print(ax1)

ax2 = plt.subplot(2,1,2)
plt.plot(np.arange(10), np.arange(10) * 2, 'r.-')
plt.title('a tale of 2 subplots')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot(range(-20, 80), np.random.randn(100).cumsum(), 'r--')
plt.title('한글 점검');
```

```python
fig = plt.figure(figsize=(15, 12))
fig, ax = plt.subplots(subplot_kw=dict(projection='3d'),
figsize=(15, 12))
graph_title = '테스트'
ax.set_title(f'{graph_title}')
ax.set_xlabel('주파수 (Hz)', labelpad=20)
ax.set_ylabel('초 (s)', labelpad=20)
ax.set_zlabel('강도/주파수 (dB/Hz)', labelpad=20)
ax.view_init(20, 20)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 8))
ax1 = fig.add_subplot(2,2,1)
plt.plot(np.random.randn(50).cumsum(), "g--")

ax2 = fig.add_subplot(2,2,2)
plt.plot(np.random.randn(50).cumsum(), "r--")

ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(), "m--")

ax4 = fig.add_subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(), "c--")
```
